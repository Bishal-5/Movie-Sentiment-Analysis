```python
# splitting text into smaller units, such as words, sentences, or n-grams
from nltk.tokenize import TweetTokenizer # For tokenizing tweets
```

```python
english_stopwords = stopwords.words('english') # Get the list of stopwords for English
hinglish_stopwords = stopwords.words('hinglish') # Get the list of stopwords for Hinglish
```

```python
def clean_tweet(tweet):
    # Remove URLs, hashtags, mentions, and special characters
    tweet = re.sub(r"http\S+|www\S+|@\w+|#\w+", "", tweet)
    tweet = re.sub(r"[^\w\s]", "", tweet)

    # Tokenize the tweet
    tokenizer = TweetTokenizer(preserve_case=False, reduce_len=True, strip_handles=True)
    tokens = tokenizer.tokenize(tweet)

    # Remove stopwords for English and Hinglish
    tokens = [token for token in tokens if token not in english_stopwords and token not in hinglish_stopwords]

    # Remove punctuation and convert to lowercase
    tokens = [token.translate(str.maketrans('', '', string.punctuation)) for token in tokens]
    tokens = [token.lower() for token in tokens]

    # Join tokens back into a string
    cleaned_tweet = ' '.join(tokens)

    return cleaned_tweet
```

```python
# Apply the clean_tweet function to each row of the 'Tweets' column
df['Cleaned_Tweets'] = df['Tweets'].apply(clean_tweet)
```

```python
def clean_text(text):
    text = text.lower() # Convert text to lowercase
    return text.strip() # Remove leading and trailing whitespaces
```

```python
def tokenization(text):
    tokens = re.split('W+',text) # Split text into words
    return tokens
```

```python
# Apply the 'tokenization' function to each tweet in the 'Cleaned_Tweets' column
df.Cleaned_Tweets = df.Cleaned_Tweets.apply(lambda x: tokenization(x))
```

```python
from nltk.stem import WordNetLemmatizer # For lemmatizing words
wordnet_lemmatizer = WordNetLemmatizer() # Create an instance of WordNetLemmatizer
```

```python
nltk.download('wordnet') # Download the WordNet lemmatizer
nltk.download('omw-1.4') # Download the WordNet lemmatizer
```

```python
def lemmatizer(text):
  # Lemmatize each word in the text
    lemm_text = "".join([wordnet_lemmatizer.lemmatize(word) for word in text])
    return lemm_text
```

```python
# Apply the 'lemmatizer' function to each tweet in the 'Cleaned_Tweets' column
df.Cleaned_Tweets = df.Cleaned_Tweets.apply(lambda x: lemmatizer(x))
```

```python
def remove_digits(text):
    clean_text = re.sub(r"\b[0-9]+\b\s*", "", text) # Remove digits
    return(text)
```

```python
def remove_digits1(sample_text):
  # Remove digits
    clean_text = " ".join([w for w in sample_text.split() if not w.isdigit()])
    return(clean_text)
```

```python
# Function to label sentiment based on polarity score
def label_sentiment(x:float):
    if x < -0.05 : return 'negative'
    if x > 0.35 : return 'positive'
    return 'neutral'
```

```python
sia = SIA() # Create an instance of the SentimentIntensityAnalyzer
```

```python
# Calculate sentiment for each tweet
df1['sentiment'] = [sia.polarity_scores(x)['compound'] for x in tqdm(df1['english_tweets'])]
# Apply the function to the 'sentiment' column
df1['overall_sentiment'] = df1['sentiment'].apply(label_sentiment);
```
